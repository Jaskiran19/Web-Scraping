```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Place URL Of Search

```python
URL = "https://www.amazon.com/s?k=shoes&crid=XYXGMHDGU79E&sprefix=shoes%2Caps%2C713&ref=nb_sb_noss_2"
```

# Place User-Agent Inside Header Passed(What Is My Browser?)

```python
HEADERS = ({'User-Agent': 'https://explore.whatismybrowser.com/useragents/parse/?analyse-my-user-agent=yes', 'Accept-Language': 'en-US, en;q=0.5' })
```

# Send Request

```python
webpage = requests.get(URL, headers=HEADERS)
```

# Recieve Response

```python
webpage 
```

```python
webpage.content
```

```python
type (webpage.content)
```

# Get The Information In HTML Format

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

# Acquire Information Specific To Anchor Or Span Tags.

```python
links = soup.find_all("a", attrs={'class' : 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```
